```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_user = pd.read_csv("/content/drive/MyDrive/Data Science Project/Social Graph PPKM/node.csv")
df_user
```

## PPKM Pro 

```python
df_user_pro = df_user.loc[df_user["stance"]=="pro"]
df_user_pro
```

```python
jml_user_pro_m = df_user_pro.gender.value_counts().m
jml_user_pro_w = df_user_pro.gender.value_counts().w
data = [jml_user_pro_m, jml_user_pro_w]
label = ["Man","Woman"]
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = label)
 
# show plot
plt.show()
```

## PPKM Kontra

```python
df_user_kontra = df_user.loc[df_user["stance"]=="kontra"]
df_user_kontra
```

```python
jml_user_kontra_m = df_user_kontra.gender.value_counts().m
jml_user_kontra_w = df_user_kontra.gender.value_counts().w
data = [jml_user_kontra_m, jml_user_kontra_w]
label = ["Man","Woman"]
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = label)
 
# show plot
plt.show()
```

## All Gender Analysis

```python
import numpy as np
outer = [df_user.gender.value_counts().m,df_user.gender.value_counts().w]
outer_label = ["Man","Woman"]

inner = np.array([[jml_user_kontra_m, jml_user_pro_m], [jml_user_kontra_w, jml_user_pro_w]])
inner_label = ["Kontra","Pro","Kontra","Pro"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(24,12))
size = 0.3

ax.pie(outer, radius=1,
       labels=outer_label,
       autopct='%1.1f%%',
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(inner.flatten(), radius=1-size, 
       labels = inner_label,
       wedgeprops=dict(width=size, edgecolor='w'))

plt.show()
```
