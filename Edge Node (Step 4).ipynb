```python
#Edge
df_user_reply = pd.read_csv("/content/drive/MyDrive/Data Science Project/Social Graph PPKM/data_user_cleaned_with_user_text_only.csv")
```

```python
df_user_reply = df_user_reply[["user_screen_name","in_reply_to_screen_name"]]
```

## Delete Target NaN

```python
df_user_reply = df_user_reply[df_user_reply.in_reply_to_screen_name.notnull()]
df_user_reply 
```

## Delete User Target Not in Data Node

```python
user_not_in_data = []
for i in df_user_reply["in_reply_to_screen_name"]:
  if df_user['user_screen_name'].str.contains(i).any():
    continue
  else:
    user_not_in_data.append(i)

user_not_in_data
```

```python
df_user_reply = df_user_reply[df_user_reply.in_reply_to_screen_name.isin(user_not_in_data) == False]
```

## Delete User Source not in Data Node

```python
source_not_in_data = []
for i in df_user_reply["user_screen_name"]:
  if df_user['user_screen_name'].str.contains(i).any():
    continue
  else:
    source_not_in_data.append(i)

source_not_in_data
```

```python
df_user_reply = df_user_reply[df_user_reply.user_screen_name.isin(source_not_in_data) == False]
```

## Delete User Node not in Edge Source and Target

```python
id_not_in_data = []
for i in df_user["user_screen_name"]:
  if (df_user_reply['user_screen_name'].str.contains(i).any())|(df_user_reply['in_reply_to_screen_name'].str.contains(i).any()):
    continue
  else:
    id_not_in_data.append(i)

id_not_in_data
```

```python
df_user = df_user[df_user.user_screen_name.isin(id_not_in_data) == False]
```

## Delete User Edge Relation to Itself
