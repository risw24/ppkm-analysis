## Data Preprocessing

1. Data to be processed is username twitter, real name user, reply to who (username) and tweet.

```python
data_user = ppkm[['user_screen_name', 'user_description', 'user_followers_count', 'user_location', 'user_name', 'user_urls','in_reply_to_screen_name', 'text']]
data_user
```

2. We only use "real people" username twitter, not the organization or business account. But, in this case, we are not using account detection to separate it, we only detect from url column, if account has url, it indicates the account is organization or business account.

```python
data_user = data_user[data_user['user_urls'].isna()]
```

```python
data_user.drop(['user_urls'], axis=1, inplace=True)
```

3. Check bot account with followers, description and location. The real account is account has big follower with description and location. (This is not really accurate)

```python
data_user = data_user.drop(data_user[(data_user.user_followers_count < 1000)].index) 
data_user
```

```python
data_user = data_user[data_user['user_description'].notna()]
data_user = data_user[data_user['user_location'].notna()]
```

```python
data_user = data_user[['user_screen_name', 'user_name','in_reply_to_screen_name', 'text']]
data_user
```

4. Remove duplicate data user

```python
data_user_unique = data_user[["user_screen_name","user_name"]]
```

```python
data_user_unique[data_user_unique["user_screen_name"].duplicated()]
```

```python
data_user_unique.drop_duplicates(subset=['user_screen_name'],inplace = True)
```
